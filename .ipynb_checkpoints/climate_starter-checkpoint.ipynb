```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
import datetime as dt
```

# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
```

```python
# create engine to hawaii.sqlite
# Need to edit this since this file is in the Resources folder
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()

# reflect the tables
Base.prepare(engine, reflect = True)
```

```python
# View all of the classes that automap found
# Displays the tables within the sqlite file
Base.classes.keys()
```

```python
# Save references to each table
MeasurementTable = Base.classes.measurement
StationTable = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Precipitation Analysis

```python
# Get a list of column names and types
# Had to go back up top and import inspector
inspector = inspect(engine)
inspector.get_table_names()
columns = inspector.get_columns('measurement')
for c in columns:
    print(c['name'], c['type'])
```

```python
# Find the most recent date in the data set.
# Stores the most recent date by searching the measurement table date column for the maximum value.
LatestDate = session.query(func.max(MeasurementTable.date)).first()
LatestDate
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 
# LatestDate

# Calculate the date one year from the last date in data set.
OneYearDeltaDate = dt.date(2017, 8, 23) - dt.timedelta(days=365)
OneYearDeltaDate

# Perform a query to retrieve the data and precipitation scores
# Instructions say to get the date and prcp columns
PrecipData = session.query(MeasurementTable.date, MeasurementTable.prcp).\
    filter(MeasurementTable.date >= OneYearDeltaDate).all()

PrecipData

# Save the query results as a Pandas DataFrame and set the index to the date column
PrecipDF = pd.DataFrame(PrecipData, columns = ['Date', 'Precipitation'])
PrecipDF.head()

# Sort the dataframe by date
PrecipDFSorted = PrecipDF.sort_values('Date')
PrecipDFSorted.head()

# Use Pandas Plotting with Matplotlib to plot the data
PrecipDFSorted.plot(x = 'Date', y = 'Precipitation', rot = 45)
plt.xlabel('Date')
plt.ylabel('Precipitation (In)')
plt.title('Preciptation Volumes Over Time')
plt.show()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
PrecipDFSorted.describe()
```

# Exploratory Station Analysis

```python
# Get a list of column names and types
inspector.get_table_names()
columns = inspector.get_columns('station')
for c in columns:
    print(c['name'], c['type'])
```

```python
# Design a query to calculate the total number stations in the dataset
session.query(func.count(StationTable.station)).all()
```

```python
# Design a query to find the most active stations (i.e. what stations have the most rows?)
# List the stations and the counts in descending order.
# Adding in the station itself to the return results
session.query(MeasurementTable.station, func.count(MeasurementTable.station)).group_by(MeasurementTable.station).\
        order_by(func.count(MeasurementTable.station).desc()).all()
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.

Minimum = session.query(func.min(MeasurementTable.tobs)).\
        filter(MeasurementTable.station == "USC00519281").all()

Maximum = session.query(func.max(MeasurementTable.tobs)).\
        filter(MeasurementTable.station == "USC00519281").all()

Average = session.query(func.avg(MeasurementTable.tobs)).\
        filter(MeasurementTable.station == "USC00519281").all()

Minimum, Maximum, Average
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
OneYearDeltaDate = dt.date(2017, 8, 23) - dt.timedelta(days=365)
OneYearDeltaDate

TwelveMosTemps = session.query(MeasurementTable.date, MeasurementTable.tobs).\
        filter(MeasurementTable.date >= OneYearDeltaDate).\
        filter(MeasurementTable.station == 'USC00519281').all()

TwelveMosTemps

TwelveMosDF = pd.DataFrame(TwelveMosTemps, columns = ['Date', 'Temperature'])

# Bins by number of months in a year
TwelveMosDF.plot.hist(bins = 12)
plt.title('Count of Temperatures over Last 12 Months')
plt.xlabel('Temperatures')
plt.ylabel('Counts')
plt.tight_layout()
plt.show
```
